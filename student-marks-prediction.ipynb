**Importing the libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Importing the dataset**

```python
dataset = pd.read_csv('/kaggle/input/student-performance-predictions/student_performance.csv')
X = dataset.iloc[:,[3,4,5,6,7]].values
y = dataset.iloc[:, -1].values
```

**Label Encoding**

```python
#here we dont have to encode name and gender as it does not quite affect the model.
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
# le1=LabelEncoder()
X[:,-1]=le.fit_transform(X[:,-1])
# X[:,1]=le1.fit_transform(X[:,1])
print(X)
```

**Training the Random forest regression model on a whole dataset**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

```python
print(regressor.predict([['90','89','25','4','0']]))
```

**Predicting the test results**

```python
y_pred = regressor.predict(X)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y.reshape(len(y),1)),1))
```

**Evaluating the code performance**

```python
from sklearn.metrics import r2_score
print("Accuracy:",r2_score(y, y_pred)*100,"%")
```
